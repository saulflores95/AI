```python
def loss(T, y):
    sigma = np.multiply(T,np.log(y))
    print('sigma: ',sigma)
    return -1*np.nansum(sigma)
```

```python
target = np.array([1,0,0])
print('Target: ',target)
y = np.array([.999,0.01,0.01])
print('predicted y: ',y)
print('Error: ',loss(target,y))
```
